```
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
from qutip import *
```

```
# set up the Hamiltonian
Bz = 5
B = 50
u = 1.0
H = u * ( Bz * sigmaz() + B * sigmax() )
print(H)

psi0 = basis(2,0)
tlist = np.linspace(0, 50, 500)
result = mesolve(H, psi0,tlist,[],[])

fig, ax = plt.subplots(figsize=(8,5))
ax.plot(tlist, expect(sigmaz(), result.states), label="z")
#ax.plot(tlist, result.expect[1], label="x")
fig.legend()
fig.show()
```
